```python
# Cell 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print(" All libraries imported!")
print(f"Pandas version: {pd.__version__}")
```

```python
# Cell 2: Load the main census data
census = pd.read_csv('../data/raw/india-districts-census-2011.csv')

print(" Data loaded successfully!")
print(f"\n Shape: {census.shape}")
print(f" Total rows: {census.shape[0]}")
print(f" Total columns: {census.shape[1]}")
```

```python
# Cell 3: See all column names
print(" ALL COLUMNS IN DATA:\n")
for i, col in enumerate(census.columns, 1):
    print(f"{i}. {col}")
```

```python
# Cell 7: Check missing values
print("🔍 MISSING VALUES PER COLUMN:\n")
missing = census.isnull().sum()
missing_pct = (missing / len(census)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct.round(2)
})

# Show only columns with missing values
missing_df = missing_df[missing_df['Missing Count'] > 0]

if len(missing_df) == 0:
    print(" No missing values! Great data!")
else:
    print(missing_df)
```

```python
# Cell 8: How many states?
print(f" Total Unique States: {census['State name'].nunique()}")
print(f"\n All States:")
print(census['State name'].unique())
```

```python
# Cell 9: Districts per state
districts_per_state = census['State name'].value_counts()
print(" NUMBER OF DISTRICTS PER STATE:\n")
print(districts_per_state)
```

```python
# Display all column names nicely
print(" ALL 118 COLUMNS IN DATA:\n")
print("="*60)

for i, col in enumerate(census.columns, 1):
    print(f"{i:3d}. {col}")
```

```python
# See first 5 rows of important columns
important_cols = ['State name', 'District name', 'Population', 'Male', 'Female', 'Literate']

# Check which columns exist
existing_cols = [col for col in important_cols if col in census.columns]
print(f"Available columns: {existing_cols}\n")

census[existing_cols].head(10)
```

```python
# Total population by state
state_population = census.groupby('State name')['Population'].sum().reset_index()
state_population = state_population.sort_values('Population', ascending=False)

print(" TOP 10 MOST POPULATED STATES:\n")
print(state_population.head(10).to_string(index=False))

print("\n\n BOTTOM 10 LEAST POPULATED STATES:\n")
print(state_population.tail(10).to_string(index=False))
```

```python
# Calculate literacy rate (% of population that can read/write)
census['Literacy_Rate'] = (census['Literate'] / census['Population']) * 100

# Top 10 most literate districts
top_literate = census[['State name', 'District name', 'Literacy_Rate']].sort_values('Literacy_Rate', ascending=False)

print(" TOP 10 MOST LITERATE DISTRICTS:\n")
print(top_literate.head(10).to_string(index=False))

print("\n\n BOTTOM 10 LEAST LITERATE DISTRICTS:\n")
print(top_literate.tail(10).to_string(index=False))
```

```python
# Calculate sex ratio (females per 1000 males)
census['Sex_Ratio'] = (census['Female'] / census['Male']) * 1000

# State wise sex ratio
state_sex_ratio = census.groupby('State name').agg({
    'Male': 'sum',
    'Female': 'sum'
}).reset_index()

state_sex_ratio['Sex_Ratio'] = (state_sex_ratio['Female'] / state_sex_ratio['Male']) * 1000
state_sex_ratio = state_sex_ratio.sort_values('Sex_Ratio')

print(" STATES WITH WORST SEX RATIO (Lowest females per 1000 males):\n")
print(state_sex_ratio.head(10)[['State name', 'Sex_Ratio']].to_string(index=False))

print("\n\n STATES WITH BEST SEX RATIO:\n")
print(state_sex_ratio.tail(10)[['State name', 'Sex_Ratio']].to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 15 most literate states
state_literacy = census.groupby('State name').agg({
    'Literate': 'sum',
    'Population': 'sum'
}).reset_index()

state_literacy['Literacy_Rate'] = (state_literacy['Literate'] / state_literacy['Population']) * 100
state_literacy = state_literacy.sort_values('Literacy_Rate', ascending=True)

# Create bar chart
plt.figure(figsize=(12, 8))
plt.barh(state_literacy['State name'], state_literacy['Literacy_Rate'], color='steelblue')
plt.xlabel('Literacy Rate (%)')
plt.ylabel('State')
plt.title(' Literacy Rate Across Indian States (2011 Census)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n Your first visualization is ready!")
```

```python
# HIDDEN PATTERN: Does literacy relate to better sex ratio?
state_analysis = census.groupby('State name').agg({
    'Literate': 'sum',
    'Population': 'sum',
    'Male': 'sum',
    'Female': 'sum'
}).reset_index()

state_analysis['Literacy_Rate'] = (state_analysis['Literate'] / state_analysis['Population']) * 100
state_analysis['Sex_Ratio'] = (state_analysis['Female'] / state_analysis['Male']) * 1000

# Calculate correlation
correlation = state_analysis['Literacy_Rate'].corr(state_analysis['Sex_Ratio'])

print(f"🔍 HIDDEN PATTERN DISCOVERED!\n")
print(f"Correlation between Literacy and Sex Ratio: {correlation:.3f}")

if correlation > 0.3:
    print(" Positive Pattern: Higher literacy → Better gender equality!")
elif correlation < -0.3:
    print(" Negative Pattern: Higher literacy → Worse gender ratio (Surprising!)")
else:
    print(" Weak relationship between literacy and gender ratio")
```

```python
# Find states with HIGH literacy but POOR sex ratio (Surprising!)
state_analysis_sorted = state_analysis.sort_values('Literacy_Rate', ascending=False)

print(" SURPRISING FINDING:\n")
print("States with HIGH literacy but POOR sex ratio:\n")
print("="*60)

# Top 10 literate states with their sex ratio
top_literate = state_analysis_sorted.head(15)
surprising = top_literate[top_literate['Sex_Ratio'] < 1000]

print(surprising[['State name', 'Literacy_Rate', 'Sex_Ratio']].to_string(index=False))

print("\n Insight: Even highly educated states have less than 1000 women per 1000 men!")
```

```python
# Check if religion columns exist
religion_cols = ['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']
existing_religion = [col for col in religion_cols if col in census.columns]

if existing_religion:
    # Total population by religion
    religion_total = census[existing_religion].sum()
    
    print(" RELIGIOUS POPULATION IN INDIA (2011):\n")
    for religion, count in religion_total.items():
        percentage = (count / census['Population'].sum()) * 100
        print(f"{religion:15s}: {count:>15,.0f}  ({percentage:.2f}%)")
    
    # Pie chart
    plt.figure(figsize=(10, 8))
    plt.pie(religion_total.values, 
            labels=religion_total.index,
            autopct='%1.1f%%',
            startangle=90,
            colors=['#FF9933', '#138808', '#0000FF', '#FFD700', '#FF6B6B', '#9B59B6'])
    plt.title('🇮🇳 Religious Distribution in India (2011)', fontsize=14, fontweight='bold')
    plt.show()
else:
    print("Religion columns not found. Check column names.")
```

```python
# Check internet/computer access
tech_cols = ['Households_with_Internet', 'Households_with_Computer', 'Households_with_Electric_Lighting']
existing_tech = [col for col in tech_cols if col in census.columns]

if existing_tech:
    print("💻 DIGITAL INDIA STORY (2011):\n")
    
    state_tech = census.groupby('State name')[existing_tech + ['Households']].sum().reset_index()
    
    if 'Households_with_Internet' in state_tech.columns:
        state_tech['Internet_Percentage'] = (state_tech['Households_with_Internet'] / state_tech['Households']) * 100
        
        # Top 10 digital states
        top_digital = state_tech.sort_values('Internet_Percentage', ascending=False).head(10)
        
        print(" TOP 10 DIGITALLY ADVANCED STATES (2011):\n")
        print(top_digital[['State name', 'Internet_Percentage']].to_string(index=False))
        
        print("\n Note: This is 2011 data - imagine how much has changed!")
else:
    print("Tech columns not found.")
```

```python
# Workers vs Non-workers
work_cols = ['Workers', 'Non_Workers', 'Male_Workers', 'Female_Workers']
existing_work = [col for col in work_cols if col in census.columns]

if existing_work:
    state_work = census.groupby('State name').agg({
        col: 'sum' for col in existing_work + ['Population']
    }).reset_index()
    
    if 'Workers' in state_work.columns:
        state_work['Worker_Percentage'] = (state_work['Workers'] / state_work['Population']) * 100
        
        # Top working states
        top_workers = state_work.sort_values('Worker_Percentage', ascending=False).head(10)
        
        print(" TOP 10 STATES BY WORKER PARTICIPATION:\n")
        print(top_workers[['State name', 'Worker_Percentage']].to_string(index=False))
        
        # Female participation
        if 'Female_Workers' in state_work.columns and 'Workers' in state_work.columns:
            state_work['Female_Worker_Share'] = (state_work['Female_Workers'] / state_work['Workers']) * 100
            
            print("\n\n STATES WITH HIGHEST FEMALE WORKER PARTICIPATION:\n")
            top_female_workers = state_work.sort_values('Female_Worker_Share', ascending=False).head(10)
            print(top_female_workers[['State name', 'Female_Worker_Share']].to_string(index=False))
```

```python
# SC/ST population
caste_cols = ['SC', 'ST']
existing_caste = [col for col in caste_cols if col in census.columns]

if existing_caste:
    state_caste = census.groupby('State name')[existing_caste + ['Population']].sum().reset_index()
    
    if 'SC' in state_caste.columns:
        state_caste['SC_Percentage'] = (state_caste['SC'] / state_caste['Population']) * 100
        
        print(" STATES WITH HIGHEST SC POPULATION %:\n")
        top_sc = state_caste.sort_values('SC_Percentage', ascending=False).head(10)
        print(top_sc[['State name', 'SC_Percentage']].to_string(index=False))
    
    if 'ST' in state_caste.columns:
        state_caste['ST_Percentage'] = (state_caste['ST'] / state_caste['Population']) * 100
        
        print("\n\n STATES WITH HIGHEST ST POPULATION %:\n")
        top_st = state_caste.sort_values('ST_Percentage', ascending=False).head(10)
        print(top_st[['State name', 'ST_Percentage']].to_string(index=False))
```
